```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#2.1 Implementacion y visualizacion de k-means en 2D (30 ptos)
# 5 ptos Cargar la base de datos kmeans1.csv utilizando pandas.#

x = pd.read_table('kmeans.1.csv', delimiter= ',')
x = ((x-x.min())/(x.max()-x.min()))
x.head()
```

```python
#Por medio de matplotlib (o seaborn) genere un graco de dispersion (scatter plot) de las variables A y B.

plt.plot(x['A'], x['B'], 'k.')
plt.ylabel('A')
plt.xlabel('B')
plt.show()
```

```python
#Aplique el algoritmo k-means (from sklearn.cluster import KMeans), observe y analice los outputs disponibles (centroides, clusters asignados y distancia dentro de las clases).

from sklearn.cluster import KMeans
k_means_1= KMeans(n_clusters=4)
k_means_1.fit(x)
```

```python


centroides = k_means_1.cluster_centers_
etiquetas = k_means_1.labels_

print("Centroides:", centroides)
print("Etiquetas de los clusters:", etiquetas)
```

```python
Inercia = k_means_1.inertia_   
print("Inercia (distancia intra cluster", Inercia) 
```

```python
# Ajuste k-means considerando 1,2,...10 clusters, guarde la distancia intra clases en un diccionario. Estudie la documentacion de la clase y modifique los argumentos init, n init y max it.
dist_intra_clust= {}
for k in range(1,11):
    k_means_1= KMeans(n_clusters=k, init='k-means++', n_init=20, max_iter=1000)
    k_means_1.fit(x)
    dist_intra_clust[k] = k_means_1.inertia_

    print("Distancias intra clusters):", dist_intra_clust)


# Utilizando matplotlib (o seaborn) genere un grafico que presente la distancia intra clases para cada k utilizado. A traves del criterio del codo defina el número de clusters óptimo.
plt.plot(list(dist_intra_clust.keys()), list(dist_intra_clust.values()), color='r')
plt.xlabel("Número de clusters")
plt.ylabel("Distancia intra clases")
plt.show()
```

```python
# 2.2 Analisis de clusters
# Aplicar el algoritmo k-means a la base de datos k-means2.csv. Aplique el procesamiento necesario a los datos para una correcta asignacion de clusters.

from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
from sklearn.metrics import davies_bouldin_score

x = pd.read_table('kmeans2.csv', delimiter= ',')


escalado = StandardScaler()
x_esc = escalado.fit_transform(x)

k_means_2=KMeans(n_clusters=4)
k_means_2.fit(x)

centroides_2 = k_means_2.cluster_centers_

print("Centroides escalados:", centroides_2)
```

```python
# Determine el numero de clusters adecuados con alguno de los tres metodos vistos en clases (regla del codo, Davies-Bouldin o Silhouette plot). Genere una funcion (o un conjunto de funciones) que le permita obtener el reporte de las metricas y visualizaciones.

from sklearn.datasets import make_blobs
resultados = {}
for k in range(2, 11):
    k_means_2 = KMeans(n_clusters=k, random_state=42)
    k_means_2.fit(x_esc)
    etiquetas = k_means_2.labels_
    silhouette = silhouette_score(x_esc, etiquetas)
    db_index = davies_bouldin_score(x_esc, etiquetas)
    resultados[k] = {"Silhouette": silhouette, "Davies-Bouldin": db_index}

print(resultados)
```
